```python
import numpy as np
from scipy.stats import ttest_ind

# 기존 제품과 신제품의 비거리 데이터
current_distances = np.array([264, 261, 267, 272, 258, 283, 258, 266, 259, 270, 263, 264, 284, 263, 260, 283, 255,
                             272, 266, 268, 270, 287, 289, 280, 272, 275, 265, 260, 278, 275, 281, 274, 273, 263, 275, 267, 279, 274, 276, 262])
new_distances = np.array([277, 269, 263, 266, 262, 251, 262, 289, 286, 264, 274, 266, 262, 271, 260, 281, 250, 263,
                         278, 264, 272, 259, 264, 280, 274, 281, 276, 269, 268, 262, 283, 250, 253, 260, 270, 263, 261, 255, 263, 279])

# 독립 표본 t-검정 실행
t_statistic, p_value = ttest_ind(current_distances, new_distances)

# 결과 출력
print("t-통계량:", t_statistic)
print("p-value:", p_value)

# 유의수준 설정 (0.05)
alpha = 0.05

# 가설 검증
if p_value < alpha:
    print("귀무가설 기각: 두 제품의 평균 비거리는 통계적으로 유의한 차이가 있다.")
else:
    print("귀무가설 채택: 두 제품의 평균 비거리는 통계적으로 유의한 차이가 없다.")
```

#### (c) 각 모델의 모집단 평균에 대한 95% 신뢰구간은 얼마인가? 두 신뢰구간이 겹치는가? 위의 가설검정 결과와 비교하여 서술하시오.

```python
import numpy as np
from scipy import stats


def confidence_interval(data, confidence=0.95):
    n = len(data)
    mean = np.mean(data)
    stderr = stats.sem(data)
    margin = stderr * stats.t.ppf((1 + confidence) / 2, n - 1)
    return (mean - margin, mean + margin)

current_ci = confidence_interval(current_distances)
new_ci = confidence_interval(new_distances)

print("기존 제품 평균 비거리의 95% 신뢰구간:", current_ci)
print("신제품 평균 비거리의 95% 신뢰구간:", new_ci)

# 95%신뢰구간이 겹치는 것을 확인 할 수 있다
# 두 제품의 평균 비거리에 대한 차이가 통계적으로 유의하지 않다는 것과 일치
# 가설검정 결과에서 귀무가설을 채택하여 두 제품의 평균 비거리에는 통계적으로 유의한 차이가 없다고 결론지었으며, 
# 신뢰구간 분석 결과도 이 결론과 일치하며, 이를 통해 기존 제품과 신제품의 평균 비거리 차이가 크지 않음을 확인
```

#### 4. 올해 초 13주 동안 토요일 저녁 뉴스의 시청률이 MBC방송은 29%, KBS방송은 28%, SBS 방송은25%, 기타 방송사가 18%로 기록되었다. 토요일 저녁 방송 편성이 변경되고 난 다음에 300가구를 선정하여 시청률을 조사한 결과에 따르면 MBC 95가구, KBS 70가구, SBS 89가구, 기타 방송이 46가구를 기록하였다. 유의수준 0.05 하에서 시청률에 변경이 있었다고 할 수 있는지 통계적 판단을 하시오.

```python
# 귀무가설(H0): 토요일 저녁 방송 편성 변경 전과 후의 시청률에 차이가 없다.
# 대립가설(H1): 토요일 저녁 방송 편성 변경 전과 후의 시청률에 차이가 있다.

from scipy.stats import chi2_contingency

# 변경 전 시청률과 변경 후 시청률
households = np.array([95, 70, 89, 46])
broadcaster = np.array([0.29, 0.28, 0.25, 0.18])

# 기대 빈도수 계산
total_households = 300
expected_counts = total_households * broadcaster

# 카이제곱 검정 수행
chi_square_statistic, p_value, _, _ = chi2_contingency(
    np.array([households, expected_counts]))

# 결과 출력
print("카이제곱 통계량:", chi_square_statistic)
print("p-value:", p_value)

# 유의수준 설정 (0.05)
alpha = 0.05

# 가설 검증
if p_value < alpha:
    print("귀무가설 기각: 토요일 저녁 방송 편성 변경 전과 후의 시청률에 차이가 있다.")
else:
    print("귀무가설 채택: 토요일 저녁 방송 편성 변경 전과 후의 시청률에 차이가 없다.")
```

```python
data = [2, 4, 6, 8, 10]

# 평균 계산
mean = sum(data) / len(data)

# 분산 계산
var = sum((i - mean) ** 2 for i in data) / len(data)

# 결과 출력
print("평균: ", mean)
print("분산: ", var)
```
