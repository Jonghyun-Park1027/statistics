```python
import numpy as np
import matplotlib.pyplot as plt

pop_mean = 0
pop_std = 16
pop_size = 900

sample_size = 1000

samples = np.random.normal(loc=pop_mean, scale=pop_std, size=(sample_size, pop_size))
sample_means = np.mean(samples, axis=1)

binom = np.random.binomial(n=10, p=0.5, size=1000)
binom_counts, binom_bins = np.histogram(binom, bins=range(12))

plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.hist(sample_means, bins=30)
plt.title('Histogram of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.subplot(122)
plt.bar(binom_bins[:-1], binom_counts, width=0.8)
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')

plt.show()
```
